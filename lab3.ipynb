## Lab 3 - Bài 3

```python
# Code lab 3  bài 3 ở đây
while True:
    n = input("nhập số nguyên dương n: ")
    if n == 'q':
        print("thoát chương trình")
        break

    try:
        n = int(n)
    except ValueError:
        print("vui lòng nhập số nguyên dương hoặc 'q' để thoát")
        continue 
    ok = True
    i = 2
    while i <(n-1):
        if n % i == 0:
            ok = False
            break
        i += 1
    if ok and n > 1:
        print(f"{n} là số nguyên tố")
    else:
        print(f"{n} không phải là số nguyên tố")
        


```

## Lab 3 - Bài 5: bài tập làm thêm

```python
# code lab 3  bài làm thêm ở đây 
#xây dựng menu chương trình máy tính đơn giản
Menu = """
        "1": Phep tinh co ban: Cong, tru, nhan, chia
        "2": Luy thua: Tinh x^y
        "3": Can bac hai: Tinh √x
        "4": Ham luong giac: Sin, cos, tan
        "5": Logarit: Log co so 10, ln, log co so tuy chon
        "6": Giai phuong trinh bac nhat
        "7": Giai phuong trinh bac 2
        "8": Lich su: Luu va xem lai cac phep tinh da thuc hien
        "9": Thoi gian: Hien thi thoi gian hien tai
        "10": Giao dien than thien: Menu de su dung bang tieng Viet"""
lich_su = []
while True: 
    print(Menu)
    lua_Chon = input("Bạn chọn chức năng :")
    match lua_Chon:
        case "0":
            print("Thoát chương trình")
            break
        case "1":
            print("Phép tính cơ bản :")
            number1 = float(input("nhập số thứ nhất: "))
            number2 = float(input("nhập số thứ hai: "))
            operator = input("nhập phép toán (+, -, *, /): ")
            if operator == "+":
                result = number1 + number2
                print(f"kết quả là: {number1} + {number2} = {result} ")
            if operator == "-":
                result = number1 - number2
                print(f"kết quả là: {number1} - {number2} = {result} ")
            if operator == "*":
                result = number1 * number2
                print(f"kết quả là: {number1} * {number2} = {result} ")
            if operator == "/" and number2 != 0:
                result = number1 / number2
                print(f"kết quả là: {number1} / {number2} = {result} ")
            
            lich_su.append(f"{number1} {operator} {number2} = {result}")      
        case "2":
            print("Lũy Thừa")
            import math
            x = float(input("nhap so x: "))
            y = float(input("nhap so y: "))
            luy_thua = x ** y
            print("lũy thừa là:", luy_thua)
            lich_su.append(f"{x}^{y} = {luy_thua}")
        case "3":
            print("Can bac hai")
            import math
            x = float(input("nhap so x: "))
            can_bac_hai = math.sqrt(x)
            print("căn bậc hai là:", can_bac_hai)
            lich_su.append(f"√{x} = {can_bac_hai}")
        case "4":
            print("Ham luong giac")
            import math
            sin = float(input("nhap so sin: "))
            cos = float(input("nhap so cos: ")) 
            tan = float(input("nhap so tan: "))
            sin = math.sin(sin)
            cos = math.cos(cos)
            tan = math.tan(tan)
            print("sin là:", sin)
            print("cos là:", cos)
            print("tan là:", tan)
            lich_su.append(f"sin = {sin}, cos = {cos}, tan = {tan}")
        case "5":
            print("Logarit")
            import math
            x = float(input("nhap so x: "))
            logarit = math.log10(x)
            print("logarit là:", logarit)
            lich_su.append(f"log10({x}) = {logarit}")

        case "6":
            print("Giai phuong trinh bac nhat")
            a = float(input("nhập số a là: "))
            b = float(input("nhập số b là: "))
            if a == 0:
                if b == 0:
                    print("phương trình vô số nghiệm")
                else:
                    print("phương trình vô nghiệm")
            else:
                x = -b/a
                print("Vậy nghiệm của phương trình là: x = ",x)
            lich_su.append(f"{a}x + {b} = 0, x = {x}")
        case "7":
            print("Giai phuong trinh bac 2")
            import math
            a = float(input("nhập số a là: "))
            b = float(input("nhập số b là: "))
            c = float(input("nhập số c là: "))

            if a == 0:
                if b == 0:
                    if c == 0:
                        print("phương trình vô số nghiệm")
                    else:
                        print("phương trình vô nghiệm")
                else:
                    x = -c/b
                    print("phương trình có một nghiệm x = ",x)
                
            if a != 0:
                delta = b**2 - 4*a*c
                if delta < 0:
                    print("phương trình vô nghiệm")
                elif delta == 0:
                    x = -b/(2*a)
                    print("phương trình có nghiệm kép x1 = ",x)
                elif delta > 0:
                    x1 = (-b + math.sqrt(delta)) / (2*a)
                    x2 = (-b - math.sqrt(delta)) / (2*a)
                    print("phương trình có hai nghiệm phân biệt x1=",x1,"và x2=",x2)
                    lich_su.append(f"{a}x^2 + {b}x + {c} = 0, x1 = {x1}, x2 = {x2}")        
        case "8":
            print("Lich su: Luu va xem lai cac phep tinh da thuc hien")
            if not lich_su:
                print("Chưa có phép tính nào được thực hiện.")
            else:
                print("Lịch sử các phép tính đã thực hiện:")
                for item in lich_su:
                    print(item)
        case "9":
            print("Thoi gian: Hien thi thoi gian hien tai")
        case "10":
            print("Giao dien than thien: Menu de su dung bang tieng Viet")
        case _:
            print("Lựa chọn không hợp lệ")
```
